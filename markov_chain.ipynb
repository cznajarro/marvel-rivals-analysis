```python
import pandas as pd

df = pd.read_csv("data/match_history.csv")#loading the data
sequence = df["result"].tolist()

transitions = {"WW": 0, "WL": 0, "LW": 0, "LL": 0}#initial counts of consecutive wins/losses
for i in range(len(sequence) - 1):
    pair = sequence[i] + sequence[i+1]
    transitions[pair] += 1

winFirst = transitions["WW"] + transitions["WL"]#chance after a win
loseFirst = transitions["LW"] + transitions["LL"]#chance after a loss

print(f"P(WW): {round(transitions['WW'] / winFirst * 100, 1)}%")
print(f"P(WL): {round(transitions['WL'] / winFirst * 100, 1)}%")
print(f"P(LW): {round(transitions['LW'] / loseFirst * 100, 1)}%")
print(f"P(LL): {round(transitions['LL'] / loseFirst * 100, 1)}%")
```
